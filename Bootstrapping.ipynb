# Standard Normal Distribution

Below we generate 1000 samples from a standard normal distribution.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


np.random.seed(123)
X = np.random.normal(loc=0, scale=1, size=30)
fig, ax = plt.subplots()
sns.kdeplot(X, ax=ax, bw_method=0.5).set_title('Histogram of Standard Normal Distribution Sample')
ax.set_xlim(-5, 5)
plt.show()
```

# Bootstrap Sampling

Bootstrapping is a statistical technique [developed in 1979](https://statistics.stanford.edu/about/history/software-tools) for the purpose of non-parametric estimation. Parameter here [refers to any summary number](https://online.stat.psu.edu/statprogram/reviews/statistical-concepts/terminology), like an average or percentage, that describes the entire population. Non-parametric statistics allows for inference about population parameters without making an assumption about the underlying distributions of the populations/parameters.

## What exactly is it?
Sampling, with replacement, from your sample. Below we sample with replacement from the above sample, X. Note that this produces a similar distribution to the original sample, albeit slightly different.

```python
fig, ax = plt.subplots()
x = np.random.choice(X, size=50, replace=True)
sns.kdeplot(x, ax=ax, bw_method=0.5).set_title('Density Plot of Bootstrap Sample, B = 50')
ax.set_xlim(-5, 5)
plt.show()
```

# Large number of boostrap samples
Let's do a bunch of bootstrap samples from this sample. Looks a lot like the density plot of the original sample.

```python
fig, ax = plt.subplots()
x = np.random.choice(X, size=50000, replace=True)
sns.kdeplot(x, ax=ax, bw_method=0.5).set_title('Density Plot of Bootstrap Samples, B = 50,000')
ax.set_xlim(-5, 5)
plt.show()
```

# This technique works for any distribution

Let's try it out for a Gamma Distribution

```python
X = np.random.gamma(shape=20, scale=1.0, size=30)
fig, ax = plt.subplots()
sns.kdeplot(X, ax=ax).set_title('Histogram of Gamma Distribution Sample')
plt.show()
```

```python
fig, ax = plt.subplots()
x = np.random.choice(X, size=50, replace=True)
sns.kdeplot(x, ax=ax).set_title('Density Plot of Gamma Bootstrap Sample, B = 50')
plt.show()
```
